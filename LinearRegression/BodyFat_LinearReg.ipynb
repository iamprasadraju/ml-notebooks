```python
# importing dataset
import kagglehub
path = kagglehub.dataset_download("fedesoriano/body-fat-prediction-dataset")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression


df = pd.read_csv(path + "/bodyfat.csv")
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['BodyFat'])
y = df['BodyFat']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

r_squared_score = model.score(X_test, y_test)
print(f"R-squared score: {r_squared_score}")
```
